```python
def ma_fonction():
    print("avant")
    yield "bonjour"      # pause ici, retourne "bonjour"
    print("après") 

def avec_return():
    print("avant")
    return "bonjour"     # s'arrête définitivement ici
    print("après")  
```

```python
ma_fonction()
avec_return()
```

```python
patients = [
    {'id': 1, 'nom': 'Dupont', 'age': 45, 'pathologie': 'diabete', 'nb_consultations': 3},
    {'id': 2, 'nom': 'Martin', 'age': 67, 'pathologie': 'hypertension', 'nb_consultations': 5},
    {'id': 3, 'nom': 'Durand', 'age': 34, 'pathologie': 'diabete', 'nb_consultations': 2},
    {'id': 4, 'nom': 'Bernard', 'age': 78, 'pathologie': 'arthrose', 'nb_consultations': 7},
    {'id': 5, 'nom': 'Petit', 'age': 52, 'pathologie': 'hypertension', 'nb_consultations': 4},
]
```

Tâches :
a) Extraire tous les patients diabétiques de plus de 40 ans
b) Calculer l'âge moyen des patients par pathologie
c) Trouver le patient avec le plus de consultations
Attendu : Code Python clair avec commentaires

```python
# a ) 
patients_a =[ patient for patient in patients if patient['pathologie'] == 'diabete' and patient['age'] > 40 ]
patients_a
```

```python
#b ) Calculer l'âge moyen des patients par pathologie
import pandas as pd
df = pd.DataFrame(patients)
df.head()
df.groupby('pathologie').age.agg(age_moyen='mean')
```

```python
#c )
df[max(df['nb_consultations']) == df['nb_consultations']]
```

```python
# a bis )
df[(df["pathologie"] == "diabete") & (df["age"] > 40)]
```

```python
import re
def url2id(video_url:str) -> str:
    """ Convertit une URL YouTube en ID de vidéo si l'URL est valide.
    Args:
        video_url (str): L'URL de la vidéo YouTube
    Returns:
        str: L'ID de la vidéo YouTube"""

    # pattern = r'^(https:\/\/)?(www\.)?(youtu.?be\.com\/(watch\?v=|embed\/|v\/)|youtu\.be\/)([\w-]{11})(\S+)?$'
    pattern = r'^(https:\/\/youtu.be/).{11}$'
    if re.match(pattern, video_url) :
        print("Valid YouTube URL")
        video_id = video_url[-11:]
        return video_id
    else:
        raise ValueError('Error : invalid YouTube URL')
```

```python
url = "https://youtu.be/68QYq9jcEIQ"
print(url[-11:])
```
